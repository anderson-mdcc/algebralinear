```python
class MetodoPotenciaInversa(object):
    def __init__(self, a, b, it):
        self.a = a
        self.x = x
        self.it = it
            
    def imprimeMatriz(self, msg, a, is_2d=False):
      print(msg)
      for l in range(0, len(a)):
          print('[', end='')
          if is_2d:
              for c in range(0, len(a[0])):
                  print(a[l][c], end=' ')
          else:
            print(a[l][0], end=' ')
          print(']')
            
      print("*"*49, '\n')
    
    
    def executaMPI(self):
        maiorValor = None
        
        for i in range(self.it):
            print("*"*49, '\n')
            print("Iteração número {}".format(i))
            d = Decomposicao(self.a, self.x)
            d.iniciaDecomposicao()
            y = np.linalg.solve(d.l, self.x)
        
            produtoAX = np.linalg.solve(d.u, y)
            self.imprimeMatriz("Produto A x", produtoAX)
            maiorValor = max(produtoAX.min(), produtoAX.max(), key=abs)
            self.x = 1.0 / maiorValor * produtoAX
            self.imprimeMatriz("Resultado de x", self.x)
    
    def calculaAutovalor(self):
        produto = np.dot(self.a, self.x)
        print(produto)
        autovalor = produto[0]/self.x[0]
        print("Autovalor encontrado = {}".format(autovalor))
    
     
    def iniciaMP(self):
        self.executaMPI()
        self.calculaAutovalor()
```

```python
print("*"*15, "Método da Potência Inversa", "*"*15)

a = np.array([[2.0, 1.0, 0.0], [2.0, 5.0, 3.0], [0.0, 1.0, 6.0]])
x = np.array([[1.0], [1.0], [1.0]])
    
mpi = MetodoPotenciaInversa(a, x, 5)
mpi.imprimeMatriz("Matriz Original", mpi.a, True)
```
